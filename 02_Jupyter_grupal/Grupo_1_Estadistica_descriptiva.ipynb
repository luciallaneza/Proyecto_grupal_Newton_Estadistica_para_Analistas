#  🎓Conceptos clave 

Ya que los tres primero conceptos se dieron en una píldora explicativa recientemente, nos vamos a centrar en los siguientes:    

 

### 📏 Varianza

   - Mide cuánto se alejan los datos de la media, por término medio. Cuanto más grande, más dispersos están los datos. Cuanto más pequeña, más juntos.   

###  📐 Desviación típica
    
   - Mide cuánto se alejan los datos de la media, en las mismas unidades que los datos. Si tus datos son euros, la desviación típica también sale en euros — y eso la hace mucho más fácil de interpretar.  
    **La varianza es la herramienta técnica. La desviación típica es la traducción humana de esa herramienta.**

###  📈 Cuartiles 
    
   - Los cuartiles son tres percentiles especiales que dividen los datos en cuatro partes iguales.
   -  Ejemplo: 
	   - Q1 = percentil 25 → por debajo está el 25% de los datos 
	   - Q2 = percentil 50 → es la mediana, justo la mitad 
	   - Q3 = percentil 75 → por debajo está el 75% de los datos.   


###  📉 Percentiles
    
   - Nos indica qué porcentaje de datos quedan por debajo del tuyo. Ejemplo:
    "Si sacas un 8 en un examen y estás en el percentil 90, significa que el 90% de tus compañeros han sacado menos que tú. No importa el número en sí — importa dónde estás respecto al grupo."

      **El cuartil/percentil mide la posición que tiene un dato respecto del conjunto total de datos**


## Consumo de la Api pública y creación del Dataframe

```python
#importar librerias 
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

#consumo api
from fastapi import FastAPI
import requests
app = FastAPI()

BASE_URL = "https://restcountries.com/v3.1/all?fields=name,capital,currencies,languages,region,population,area,density"
response = requests.get(BASE_URL, timeout=20)
response.raise_for_status()
data = response.json()
print(data)

#dataframe
rows = []
for c in data:#aquí limitamos las columnas que queremos visualizar de las que hemos consumido en la api.
    rows.append({
        "country": c.get("name", {}).get("common"),
        "region": c.get("region"),
        "population": c.get("population"),
        "area": c.get("area"),
        "languages": c.get("languages"),
        "currencies": c.get("currencies"),
        
    })

df = pd.DataFrame(rows)  #llamar a esa variable y convertirla en formato tabla.
# Limpieza minima para el analisis
numeric_cols = ["population", "area"] #Voy a sobrescribir las columnas originales con la versión nueva y limpia.
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce") #Esto es lo que evita que el código se rompa si encuentra basura. Si encuenta un texto como numero lo convierte a numero.
#Con estas dos líneas, obligas a que los datos se conviertan en números de verdad para poder hacer cálculos (sumar, promediar, etc.).
df = df.dropna(subset=["country", "region", "population", "area"]) #elimina los numeros nulos 
df = df[df["population"] > 0] #valores mayores de cero
df = df[df["area"] > 0]
df_europa = df[df["region"] == "Europe"].copy() #para filtrar las regiones.

display(df_europa) #visualizar una tabla numpy.

# Variable derivada util para analisis
df["density"] = df["population"] / df["area"]
```

```python
#Media de la población en Europa
print(f"Poblacion media: {df['population'].mean():,.0f}")
#Mediana de la población en Europa
print(f"Poblacion mediana: {df['population'].median():,.0f}")
#Moda de Lenguaje en Europa
print(f"Moda sobre el lenguage en europa: {df['languages'].mode()[0]}")
#Varianza de la población en Europa 
print(f"Varianza de la población: {df["population"].var(ddof=0)}")
#Desviación típica de la población en Europa
print(f"Desviación típica: {df['population'].std(ddof=0)}")
#Cuartiles sobre el Area en Europa
print(f"Los cuartiles son:\n{df["area"].quantile([0.25, 0.5, 0.75])}") 
#Percentiles sobre el Area en Europa 
print(f"Los percentiles son:\n{df["area"].quantile([0.9])}")
#Total de paises en Europa
print(f"Total de países en Europa: {len(df[df['region'] == 'Europe'])}")

##:,.0f s un formateador de cadenas (f-string).
#  Sirve para que los números grandes no se vean como un desorden de decimales, sino como una cifra profesional y fácil de leer.
## ddof=0 para población
## Al poner [0], le estás diciendo a Python: "De todos los posibles ganadores que me devuelva la moda, dame solo el primero". 
# Con el [0]: Extraes el texto puro.
```
